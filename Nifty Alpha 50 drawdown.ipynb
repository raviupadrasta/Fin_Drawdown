```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Replace 'nifty50.csv' and 'niftyalpha50.csv' with your actual file paths
nifty50 = pd.read_csv('nifty50.csv', index_col='Date', parse_dates=True)
nifty50.sort_values(by=['Date'],ascending=True)
nifty_alpha50 = pd.read_csv('niftyalpha50.csv', index_col='Date', parse_dates=True)
nifty_alpha50.sort_values(by=['Date'],ascending=True)
```

```python
def calculate_drawdown(series):
  """Calculates the drawdown of a given series."""
  
  # Calculate the cumulative maximum
  cum_max = series.cummax()
  print(cum_max)
  # Calculate the drawdown from the peak
  drawdown = (series - cum_max) / cum_max
  
  return drawdown

#nifty50_drawdown = calculate_drawdown(nifty50['Close '])
nifty_alpha50_drawdown = calculate_drawdown(nifty_alpha50['Close'])

```

```python
fig, ax1 = plt.subplots(figsize=(12, 6))

# Plot the price series
ax1.plot(nifty50['Close'], label='Nifty 50', color='blue')
ax1.plot(nifty_alpha50['Close'], label='Nifty Alpha 50', color='orange')
ax1.set_xlabel('Date')
ax1.set_ylabel('Index Value', color='blue')
ax1.tick_params(axis='y', labelcolor='blue')
ax1.legend(loc='upper left')


plt.grid(True)
plt.show()
```

```python
# Plot the drawdown
fig, ax2 = plt.subplots(figsize=(12, 6))

#ax2 = ax1.twinx()
#ax2.plot(nifty50_drawdown, label='Nifty 50 Drawdown', color='red', linestyle='--')
ax2.plot(nifty_alpha50_drawdown, label='Nifty Alpha 50 Drawdown', color='green', linestyle='-')
ax2.set_ylabel('Drawdown', color='green')
ax2.tick_params(axis='y', labelcolor='green')
ax2.legend(loc='lower right')

plt.title('Nifty Alpha 50 with Drawdown')
plt.grid(True)
plt.show()
```
